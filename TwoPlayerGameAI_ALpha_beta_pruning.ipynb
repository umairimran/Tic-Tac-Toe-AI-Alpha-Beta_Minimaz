```python
"""
Tic Tac Toe Player
"""

import copy
import math
import random
AI = "X"
Human = "O"
Draw="D"
EMPTY = None
def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]
def player(board):
    """
    Returns player who has the next turn on a board. based on even odd value
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return AI
    return Human
def actions(board):
    """
    just check if a cell is empty and can be made a move over.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def print_board(board):
    for row in board:
        print(" | ".join([str(x) if x is not None else " " for x in row]))
        print("-" * 9)
    print()
def result(board, action):
    """
    places curr players symbol over the board
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

```

```python
def terminal(board):
    """
    Returns True if game is over or not
    """
    if winner(board) != None:
        return True

    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == AI:
        return 1
    elif winner_val == Human:
        return -1
    return 0
```
